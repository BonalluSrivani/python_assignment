### Question 3: Write a Python program to check if a string is a palindrome don't use [::-1].

- `sample input : `
    - `var = 'NUN'`
- `sample output :`
    - `result : It is a Palindrome`

```python
var = 'NUN'
n = ''
for i in var:
  n = i+n
if var==n:
  print(f'{var}->it is palindrome')
else:
  print(f'{var}->its not a palindrome')


```

### Question 5: Write a Python program to check if two strings are anagrams.

- `sample input : `
    - `var1 = 'listen'`
    - `var2 = 'silent'`
- `sample output :`
    - `result : both are anagram because characters in var2 is in var1`

```python
var1 = 'listen'
var2 = 'slient'
s = 0
for i in var1:
  if i in var2:
    s+=1
if s ==len(var1):
  print(' both are anagram because characters in var2 is in var1')
else:
  print('both are not a anagram')




```

### Question 6: Write a Python program to remove all the vowels from a string.

- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : pythn , because vowels are [AEIOUaeiou]`
    





```python
var = 'python'
s = ''
vowel='AEIOUaeiou'
for i in var:
  if i not in vowel:
    s+=i
print(f'{s} , because vowels are [AEIOUaeiou]')
```

### Question 7: Write a Python program to find the longest word in a string.

- `sample input : `
    - `var = 'python is Easy Language'`
- `sample output :`
    - `result : Language`

```python
var =  'python is Easy Language'
v = ''
for i in var.split():
  if len(i)>len(v):
    v=i
print(v)

```

### Question 9: Write a Python program to find the frequency of each character in a string.

- `sample input : `
    - `var = 'sharuk khan'`
- `sample output :`
    - `result : {'s': 1, 'h': 2, 'a': 2, 'r': 1, 'u': 1, 'k': 2, ' ': 1, 'n': 1}`

```python
var = 'sharuk khan'
d = {}
for i in var:
  if i in d:
    d[i]=d[i]+1
  else:
    d[i]=1
print(d)

```

### Question 10: write a python programme to find the sum of all the even characters based on Ascii values

- `sample input : `
    - `var = 'sharuk khan'`
- `sample output :`
    - `result : 464`


```python
var = 'sharuk khan'
s = 0
for i in var:
  if ord(i)%2==0:
    s+=ord(i)
print(s)
```
